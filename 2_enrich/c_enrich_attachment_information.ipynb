```python
import os
import json
import mimetypes

from pathlib import Path
```

```python
data_dir = "../data"

input_dir = os.path.join(data_dir, "2_enrich", "b_pictures")

output_dir = os.path.join(data_dir, "2_enrich", "c_attachment_information")
os.makedirs(output_dir, exist_ok=True)

mimetypes.init()
```

```python



for filename in os.listdir(input_dir):
    if filename.endswith('.json'):
        input_file_path = os.path.join(input_dir, filename)
        output_file_path = os.path.join(output_dir, filename)
        
        with open(input_file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        
        if 'Messages' in data:
            for message in data['Messages']:
                if 'Attachments' in message:
                    for attachment in message['Attachments']:
                        filename_for_extension = attachment.get('original_filename', attachment.get('path', ''))
                        
                        extension = Path(filename_for_extension).suffix.lower()
                        
                        mimetype, _ = mimetypes.guess_type(filename_for_extension, strict=False)
                        
                        attachment['extension'] = extension
                        attachment['mimetype'] = mimetype if mimetype else 'application/octet-stream'
        
        with open(output_file_path, 'w', encoding='utf-8') as f:
            json.dump(data, f, indent=2, ensure_ascii=False)
        
        print(f"Processed: {filename}")
```
